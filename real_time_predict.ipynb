```python
import numpy as np
import joblib
import time
from tensorflow.keras.models import load_model

model = load_model('model.h5')
scaler = joblib.load('scaler.pkl')
history = []
last_ts = None

print("Real-time prediction started...")

while True:
    try:
        line = open('stream.txt').read().strip()
        ts, val = line.split(',')

        if ts == last_ts:
            time.sleep(0.5)
            continue
        last_ts = ts

        history.append(float(val))
        if len(history) >= 24:
            input_seq = np.array(history[-24:]).reshape(-1, 1)
            input_scaled = scaler.transform(input_seq).reshape(1, 24, 1)
            pred_scaled = model.predict(input_scaled)[0][0]
            pred = scaler.inverse_transform([[pred_scaled]])[0][0]
            print(f" Next-hour forecast from {ts} → {pred:.1f} vehicles")

        time.sleep(1)
    except Exception as e:
        print("Error:", e)
        time.sleep(1)
```
